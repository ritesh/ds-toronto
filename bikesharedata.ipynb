```python
import numpy as np
import pandas as pd
# Remember to install this as well
import dateparser
```

```python
q12019 = pd.read_csv("2019/2019-Q1.csv")
```

```python
def delta(start, end):
    d = [dateparser.parse(e2) - dateparser.parse(e1) for (e1, e2) in zip(start, end)]
    return pd.Series(d)
```
